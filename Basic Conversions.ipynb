```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pylab as plt

from skimage import color
```

```python
img =mpimg.imread('C:/Users/nalindrenn/Desktop/IMG_4781.JPG')
img = np.array(img) 
plt.figure(figsize=(10,10))
imgplot=plt.imshow(img)
plt.show()
```

```python
imgGray = color.rgb2gray(img)      #store the black and white image
```

```python
plt.subplot(121), plt.imshow(img, cmap='gray')        #
plt.subplot(122), plt.imshow(imgGray, cmap='gray')

plt.show()
```

```python
im_hsv = color.rgb2hsv(img)     ##convert to hsv
plt.gray()
plt.figure(figsize=(10,8))
plt.subplot(221), plt.imshow(im_hsv[...,0]), plt.title('h',size=20), plt.axis('off')
plt.subplot(222), plt.imshow(im_hsv[...,1]), plt.title('s',size=20), plt.axis('off')
plt.subplot(223), plt.imshow(im_hsv[...,2]), plt.title('v',size=20), plt.axis('off')
plt.subplot(224), plt.axis('off')
plt.show()
```
